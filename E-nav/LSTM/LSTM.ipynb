```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import collections
import pickle
import datetime
import re
```

```python
from gensim.models import Word2Vec
```

```python
with open('train_kr_all.pkl', 'rb') as f :
        train_kr = pickle.load(f)
```

```python
word2vec_kr_model = Word2Vec.load('kr_word2vec_all.model')
```

```python
word_embeddings = word2vec_kr_model.wv.vectors
```

```python
word_vocab = word2vec_kr_model.wv.index2word
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf_vectorizer = TfidfVectorizer(vocabulary = word_vocab)
```

```python
tf_idf_kr = tf_idf_vectorizer.fit_transform(train_kr.Script.map(lambda x : ' '.join(x))).toarray()
```

```python
train_x = np.dot(tf_idf_kr, word_embeddings)
```

```python
train_kr['temp_code'] = train_kr.Script_num.map(lambda x : x[:-3])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbl = LabelEncoder()
```

```python
lbl.fit(train_kr.Category)
```

```python
lbl.classes_
```

```python
train_kr['target'] = lbl.transform(train_kr.Category)
```

```python
train_kr.reset_index(drop = True, inplace = True)
```

```python
train_kr
```

```python
from gensim.models.doc2vec import Doc2Vec
```

```python
model = Doc2Vec.load('kr_doc2vec_all.model')
```

```python
doc2vec_embedding = model.docvecs.vectors_docs
```

```python
word_mean = []
temp = []
for text in train_kr.Script :
        for word in text :
            try :
                temp.append(word2vec_kr_model[word])
            except :
                pass
        if len(temp) == 0 :
            word_mean.append(np.zeros(300).tolist())
        else :
            word_mean.append(np.array(temp).mean(axis = 0).tolist())
        temp = []
```

```python
word_mean = np.array(word_mean)
```

```python
temp_set = []
indexer = 'UL-MV-01	'
for index, row in train_kr.iterrows() :
    if indexer != row.temp_code :
        word_based_data = train_x[temp_set]
        doc_based_data = doc2vec_embedding[temp_set]
        mean_data = word_mean[temp_set]
        indexer = row.temp_code
        filename = row.temp_code
        label = row.target
        
        inputs = {'word_based_data' : word_based_data, 'doc_based_data' : doc_based_data, 'mean_data' : mean_data, 'label' : label}
        
        with open('data/{}'.format(filename), 'wb') as f :
            pickle.dump(inputs, f)
        temp_set = []
    else :
         temp_set.append(index)
```

```python
temp_dict = {}
for index, row in train_kr.iterrows() :
    if indexer != row.temp_code :
        label = row.target
        temp_dict[row.temp_code] = row.target
```

```python
with open('label_dict', 'wb') as f :
    pickle.dump(temp_dict, f)
```

```python
names = []
word_based_datas = []
doc_based_datas = []
mean_datas = []
labels = []
length = []
for file in os.listdir('data') :
    with open('data/{}'.format(file), 'rb') as f :
        temp = pickle.load(f)
    names.append(file)
    pad_len = 50 - temp['word_based_data'].shape[0]
    length.append(temp['word_based_data'].shape[0])
    word_based_datas.append(np.pad(temp['word_based_data'], [[0, pad_len], [0, 0]], 'constant', constant_values=(0)))
    doc_based_datas.append(np.pad(temp['doc_based_data'], [[0, pad_len], [0, 0]], 'constant', constant_values=(0)))
    mean_datas.append(np.pad(temp['mean_data'], [[0, pad_len], [0, 0]], 'constant', constant_values=(0)))
    labels.append(temp['label'])
```

```python
names = np.array(names)
labels = np.array(labels)
word_based_datas = np.array(word_based_datas)
doc_based_datas = np.array(doc_based_datas)
mean_datas = np.array(mean_datas)
length = np.array(length)
```

```python
train_data = {'names' : names, 'word_based_data' : word_based_datas, 'doc_based_data' : doc_based_datas, 'mean_data' : mean_datas, 'label' : labels, 'length' : length}
```

```python
with open('train_data_all.pickle', 'wb') as f :
    pickle.dump(train_data, f)
```
