```python
import pandas as pd
import numpy as np
import os
import openpyxl
import pickle
```

```python
file = os.path.join(os.getcwd(), '20170918 넘버링 울산항 전체 파일.xlsx')
```

```python
workbook = openpyxl.load_workbook(file)
```

```python
sheet = workbook.active
```

```python
matrix = []
for row in sheet.rows :
    index = row[0].row
    column = []
    for i in range(1, 14) :
        column.append(row[i].value)
    matrix.append(column)
matrix = np.array(matrix)
data = pd.DataFrame(matrix)
```

```python
data.columns = ['Port', 'Port_case', 'Category', 'Script_num', 'Port_num', 'Ship_num', 'Language', 'Script', 'Move', 'Step', 'Strategy', 'Number', 'Time']
```

```python
data = data[0:-3]
```

```python
data.drop(0, inplace = True)
```

```python
data
```

```python
data = data[data.Category != 'Communication failure(CF)']
data = data[data.Category != 'Uncategorized(UC)']
data = data[data.Category != 'Untranscriptable(UT)']
```
